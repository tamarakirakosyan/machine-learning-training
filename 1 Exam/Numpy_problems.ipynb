```python
from PIL import Image
import numpy as np
```

```python
img = np.array(Image.open("Penguins_2.jpg"))
```

** I have loaded an image into variable `img` which is a numpy array **

** Import matplotlib **

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

** Use matplotlib to visualize the `img **

```python
plt.imshow(img)
```

** Check the shape of the `img` **

```python
img.shape
```

** Check the range of numbers of the picture **
** Numbers either are from 0-255 or 0-1(normalized) **
** If the picture is in range 0-255, normalize the picture by dividing by 255 **

** Be careful to not normalize more than once **

```python
img = img/255
img[:,:,0]
```

** Find the standard deviation and mean of each channels RGB **

```python
Imean = img[:, :, 0].mean(axis = (0, 1))
Istd = img[:, :, 0].std(axis = (0, 1))
print(f"Red Channel, (std, mean) \n {Istd, Imean}")
```

```python
Imean = img[:, :, 1].mean(axis = (0, 1))
Istd = img[:, :, 1].std(axis = (0, 1))
print(f"Red Channel, (std, mean) \n {Istd, Imean}")
```

```python
Imean = img[:, :, 2].mean(axis = (0, 1))
Istd = img[:, :, 2].std(axis = (0, 1))
print(f"Red Channel, (std, mean) \n {Istd, Imean}")
```

** Mirror the image from upside down and visualize it**

```python
plt.imshow(img[::-1,::-1,:])
```

** Crop the image as close as possible to the penguins **

```python
croped_pengs = img[50:140,70:155,:]
plt.imshow(croped_pengs)
```

** By using `.transpose()` method, try to rotate the image by 90 degrees **

```python
transposed_img = np.concatenate((img[:,:,0].transpose().reshape(236,177,1), img[:,:,1].transpose().reshape(236,177,1), img[:,:,2].transpose().reshape(236,177,1)), axis=2)
plt.imshow(transposed_img)
```

** Change the order of channels from RGB to GRB **

```python
plt.imshow(img[:,:,::-1])
```

** Bonus: Change the color of the penguins' head to any color you want **
